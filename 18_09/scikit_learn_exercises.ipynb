# Exercise 1 #
- load the Iris dataset
- standardize the features using StandardScaler
- split the dataset in 70% train and 30% test
- apply a DecisionTreeClassifier
- calculate performance using classification_report (precision, recall, F1-score)
- show the confusion matrix

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = load_iris()
x = data.data
y = data.target
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.3,random_state=123)
```

```python
model = DecisionTreeClassifier(random_state=123)
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
accuracy = accuracy_score(y_test,predictions)
print(f"Accuracy = {100*accuracy:.2f}%")
```

```python
report = classification_report(y_test,predictions,target_names=data.target_names)
print(f'Classification report:\n{report}')
```

```python
conf_mat = confusion_matrix(y_test,predictions)
print(conf_mat)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues',
            xticklabels=data.target_names,
            yticklabels=data.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

# Exercise 2 #
- load the Wine dataset from `sklearn.datasets`
- standardize the features using StandardScaler
- split the dataset into 70% training and 30% testing
- apply a `DecisionTreeClassifier`
- show the `classification_report`
- show the `confusion_matrix`

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
data = load_wine()
x = data.data
y = data.target

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.3,random_state=123)
```

```python
model = DecisionTreeClassifier(random_state=123)
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
accuracy = accuracy_score(y_test,predictions)
print(f"Accuracy = {100*accuracy:.2f}%")
```

```python
report = classification_report(y_test,predictions,target_names=data.target_names)
print(f'Classification report:\n{report}')
```

```python
conf_mat = confusion_matrix(y_test,predictions)
print(conf_mat)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues',
            xticklabels=data.target_names,
            yticklabels=data.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

train_info = np.unique(y_train,return_counts=True)
plt.subplot(1,2,1)
plt.bar(train_info[0],train_info[1])
plt.xticks(train_info[0])
plt.yticks()
plt.xlim()
plt.ylim()
plt.title("Class distribution in training set")
plt.xlabel("classes")
plt.ylabel("frequency")

test_info = np.unique(y_test,return_counts=True)
plt.subplot(1,2,2)
plt.bar(test_info[0],test_info[1])
plt.xticks(test_info[0])
plt.yticks()
plt.xlim()
plt.ylim()
plt.title("Class distribution in testing set")
plt.xlabel("classes")
plt.ylabel("frequency")

plt.tight_layout()
plt.show()
```
