# Exercise 1 #
- load the Iris dataset
- standardize the features using StandardScaler
- split the dataset in 70% train and 30% test
- apply a DecisionTreeClassifier
- calculate performance using classification_report (precision, recall, F1-score)
- show the confusion matrix

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = load_iris()
x = data.data
y = data.target
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.3,random_state=123)
```

```python
model = DecisionTreeClassifier(random_state=123)
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
accuracy = accuracy_score(y_test,predictions)
print(f"Accuracy = {100*accuracy:.2f}%")
```

```python
report = classification_report(y_test,predictions,target_names=data.target_names)
print(f'Classification report:\n{report}')
```

```python
conf_mat = confusion_matrix(y_test,predictions)
print(conf_mat)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues',
            xticklabels=data.target_names,
            yticklabels=data.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
