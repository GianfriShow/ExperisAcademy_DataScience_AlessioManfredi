```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="darkgrid")  # change theme

data = np.random.normal(size=100000)  # sample from a Gaussian (standard if no parameters are specified)

sns.histplot(data, kde=True)  # plot a histogram and show the Kernel Density Estimation (estimation of the continuous distribution line)
plt.title('100k points from a standard Gaussian')
plt.show()
```

```python
fig = plt.figure()  # create a figure (canvas for a plot)
```

```python
axis = fig.add_subplot(111)  # add an axis to the figure in position 111
```

```python
# Line Graph

x = [1,2,3,4,5]
y = [2,3,5,7,11]

plt.figure()  # open the figure
plt.plot(x,y)  # plot the data points on the figure, specifying x coordinates then y coordinates
plt.title('line graph')  # add a title to the figure
plt.xlabel('x axis')  # title for the x axis
plt.ylabel('y axis')  # title for the y axis
plt.show()  # show the plot
```

```python
# Bar Plot

categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 2, 5, 8]

plt.figure()
plt.bar(categories, values)  # the method is 'bar' now
plt.title('bar plot')
plt.xlabel('categories')
plt.ylabel('values')
plt.show()
```

```python
# Histogram

data = np.random.randn(1000)  # sample from a standard Gaussian only

plt.figure()
plt.hist(data, bins=30)  # method is 'hist' for histograms
plt.title('histogram')
plt.xlabel('values')
plt.ylabel('frequency')
plt.show()
```

```python
# Scatter Plot

x = np.random.rand(50)  # sample from a standard Gaussian in a given output shape
y = np.random.rand(50)

plt.figure()
plt.scatter(x, y)
plt.title('scatter plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
